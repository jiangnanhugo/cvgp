```python

import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rc
palette=['#ff3b30', '#007aff', '#4cd964','#ffcc00', '#5856d6', '#ff9500']
markers=['o', '^', '<', 'd', '*','2']
sns.set_style("ticks")
import matplotlib as mpl
mpl.rcParams['lines.markersize'] = 6
mpl.rcParams['lines.linewidth']=1
rc("font", **{'family':'serif','serif':['Palatino'], 'size':14})
rc('text', usetex=True)
import os
import pandas as pd
```

```python
basepath="./plots"
```

```python

def plot_gp_egp(total, name, title, use_logy=False, show_ylabel=True):
    GP=[abs(float(line.split("\t")[0])) for line in total.split("\n")]
    EGP=[abs(float(line.split("\t")[1])) for line in total.split("\n")]
    noise=[abs(float(line.split("\t")[2])) for line in total.split("\n")]
    df=pd.DataFrame({
        "model": EGP+GP,
        "type": ['CVGP (ours)']*60+['GP']*60,
        "noise":noise+noise
    })
    

    plt.figure(figsize=(4.5,2.5))
    
    ax=sns.boxplot(data=df, y="model", x="noise", hue='type',
                   showfliers=True, 
                   width=0.5,
                    palette=palette)
    ax.set_title(title, fontsize=16)
    if use_logy==True:
        plt.yscale('log', base=2)
    if show_ylabel==True:
        ax.set(ylabel="NMSE", xlabel="Noise Level")
    else:
        ax.set(ylabel="", xlabel="")
    plt.legend(title="", loc='lower left')
    plt.xticks(rotation=0)
    fname = os.path.join(basepath, name+".pdf")
    plt.savefig(fname, bbox_inches='tight', pad_inches=0)
```
